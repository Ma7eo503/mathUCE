1. Conjunción (∧): "y"
2. Disyunción (∨): "o"
3. Negación (¬): "no"
4. Implicación (→): "si... entonces"
5. Doble implicación (↔): "si y solo si"

**c) Cuantificadores lógicos**

1. **∀:** Universal "Para todo"

Ejemplo: (∀x) ($x + 3 = 4$) x ∈ Z **"Falso"**

2. **∃:** Existencial "Existe"

Ejemplo: (∃x) ($x + 1/2 = 0$) **"Verdadero"**
