**Entorno de Python, variables y operadores. Principios de demostración matemática.**

**-Python**

Es un lenguaje de programación de alto nivel, interpretado y de propósito general. Fue creado por Guido van Rossum y su primera versión fue lanzada en 1991. Python es conocido por su sintaxis clara y legible, lo que facilita su aprendizaje y uso, así como por su versatilidad para distintos tipos de desarrollo, desde aplicaciones web hasta análisis de datos y aprendizaje automático.

**Características de Python**

1. **Sintaxis Sencilla y Legible:**

* La sintaxis de Python está diseñada para ser fácil de leer y escribir, lo que mejora la productividad del desarrollador.

* Usa indentación para definir bloques de código, lo que refuerza la legibilidad.

2. **Multiparadigma:**

* Soporta varios paradigmas de programación, incluidos el orientado a objetos, el imperativo, el funcional y el procedimental.

3. **Interpretado:**


* Python es un lenguaje interpretado, lo que significa que el código se ejecuta línea por línea sin necesidad de una compilación previa.

4. **Tipado Dinámico:**

* Los tipos de datos se asignan en tiempo de ejecución, lo que permite mayor flexibilidad pero puede llevar a errores si no se maneja adecuadamente.

5. **Bibliotecas Estándar Extensas:**

* Python cuenta con una amplia biblioteca estándar que facilita la realización de tareas comunes sin necesidad de instalar paquetes adicionales.

6. **Portabilidad:**

* Python puede ejecutarse en diferentes plataformas sin necesidad de modificaciones en el código (Windows, macOS, Linux, etc.).

7. **Comunidad Activa:**

* Python tiene una gran comunidad de desarrolladores que contribuyen con herramientas, bibliotecas y soporte técnico.

**Ventajas de Python**

1. **Fácil de Aprender y Usar:**

* Ideal para principiantes gracias a su sintaxis sencilla y su estructura clara.

2. **Productividad Alta:**

* La simplicidad del lenguaje permite a los desarrolladores concentrarse en resolver problemas en lugar de preocuparse por la sintaxis.

3. **Versatilidad:**

* Se puede usar para desarrollo web (Django, Flask), análisis de datos (Pandas, NumPy), inteligencia artificial (TensorFlow, PyTorch), scripting, automatización y más.

4. **Bibliotecas y Frameworks:**

* Una gran cantidad de bibliotecas y frameworks están disponibles para facilitar el desarrollo en diversas áreas.

5. **Portabilidad:**

* El mismo código puede ejecutarse en múltiples plataformas.

6. **Comunidad y Soporte:**

* La comunidad activa y los recursos disponibles (documentación, foros, cursos) facilitan la resolución de problemas y el aprendizaje continuo.

Desventajas de Python

1. **Rendimiento:**

* Como es un lenguaje interpretado, Python puede ser más lento en comparación con lenguajes compilados como C o C++. No es la mejor opción para aplicaciones que requieren un rendimiento extremadamente alto.

2. **Uso de Memoria:**

* Python tiende a usar más memoria que lenguajes como C o C++, lo cual puede ser una limitación en aplicaciones donde la gestión de recursos es crítica.

3. **No es el Mejor para Desarrollo de Apps Móviles:**

* Aunque existen frameworks como Kivy y BeeWare, Python no es tan popular ni eficiente como otros lenguajes como Java, Swift o Kotlin para el desarrollo de aplicaciones móviles.

4. **Limitaciones en Entornos de Computación de Alto Rendimiento:**

* En aplicaciones que requieren un uso intensivo de recursos (como juegos o simulaciones científicas), Python puede no ser la mejor opción sin la integración con lenguajes de bajo nivel.

5. **Errores en Tiempo de Ejecución:**

* Debido a su tipado dinámico, algunos errores solo se detectan en tiempo de ejecución, lo que puede llevar a fallos si no se manejan adecuadamente las excepciones.

**Usos Comunes de Python**

1. **Desarrollo Web:**

* Frameworks como Django y Flask permiten crear aplicaciones web robustas y escalables.

2. **Ciencia de Datos y Análisis:**

* Herramientas como Pandas, NumPy y Jupyter Notebook son ampliamente utilizadas para análisis de datos, visualización y computación científica.

3. **Inteligencia Artificial y Machine Learning:**

* Bibliotecas como TensorFlow, PyTorch y scikit-learn facilitan el desarrollo de modelos de aprendizaje automático e inteligencia artificial.

4. **Automatización y Scripting:**

* Python es popular para escribir scripts que automatizan tareas repetitivas, ahorrando tiempo y esfuerzo.

5. **Desarrollo de Juegos:**

* Frameworks como Pygame permiten la creación de juegos sencillos y prototipos.

6. **Computación en la Nube y DevOps:**

* Herramientas y scripts en Python son comunes en la administración de infraestructuras en la nube y en tareas de DevOps.

**Lenguaje de Maquina**

El lenguaje de máquina o código máquina es el sistema de códigos directamente interpretable por un circuito microprogramable, como el microprocesador de una computadora o el microcontrolador de un autómata.

**Interprete**

El intérprete ejecuta el programa directamente, traduciendo cada sentencia en una secuencia de una o más subrutinas ya compiladas en código máquina. Los términos lenguaje interpretado y lenguaje compilado​ no están bien definidos porque, en teoría, cualquier lenguaje de programación puede ser interpretado o compilado.

image.png

**Compilador**

Un compilador es un programa informático que traduce un programa escrito en un lenguaje de programación, definido «código fuente», a otro lenguaje de alto (COBOL, PASCAL, BASIC, C, etc.), medio o más bajo nivel (código intermedio o código máquina) como puede ser C/C++.

**Variables**

as variables en Python se utilizan para almacenar datos. Puedes asignar valores a las variables utilizando el operador de asignación (=). Los nombres de las variables deben comenzar con una letra o un guion bajo, y pueden contener letras, números y guiones bajos.

# Asignación de valores a variables
x = 5
y = "Hola"
z = 3.14

# Impresión de variables
print(x)  # 5
print(y)  # Hola
print(z)  # 3.14

**Reglas de Nomenclatura:**

* Los nombres de las variables deben comenzar con una letra (a-z, A-Z) o un guionbajo (_).

* El nombre de una variable no puede comenzar con un número.

* Los nombres de las variables pueden contener letras, números y guiones bajos.

* Los nombres de las variables distinguen entre mayúsculas y minúsculas (edad, Edad y EDAD son diferentes).

**Tipos de Datos:**

**Python soporta varios tipos de datos básicos:**

* **int:** Enteros, por ejemplo, x = 5

* **float:** Números de punto flotante, por ejemplo, y = 3.14

* **str:** Cadenas de texto, por ejemplo, z = "Hola"

* **bool:** Valores booleanos, True o False, por ejemplo, a = True

**Operadores en Python**

Los operadores se utilizan para realizar operaciones sobre variables y valores. Python soporta diversos tipos de operadores, incluyendo operadores aritméticos, de comparación, lógicos y de asignación.

**Operadores Aritméticos:**

(+) : Suma

(-) : Resta

(*) : Multiplicación

(/) : División

(%) : Módulo (residuo de la división)

(//) : División entera

# Ejemplos de operadores aritméticos
a = 10

b = 3

print(a + b)  # Suma: 13

print(a - b)  # Resta: 7

print(a * b)  # Multiplicación: 30

print(a / b)  # División: 3.3333333333333335

print(a % b)  # Módulo: 1

print(a ** b) # Exponenciación: 1000

print(a // b) # División entera: 3

**Operadores de Comparación:**

(==) : Igual a

(!=) : No igual a

(>) : Mayor que

(<) : Menor que

(>=) : Mayor o igual que

(<=) : Menor o igual que

# Ejemplos de operadores de comparación
x = 5

y = 3

print(x == y)  # False

print(x != y)  # True

print(x > y)   # True

print(x < y)   # False

print(x >= y)  # True

print(x <= y)  # False

**Operadores Lógicos:**

* **and:** Devuelve True si ambas condiciones son verdaderas

* **or:** Devuelve True si al menos una condición es verdadera

* **not:** Invierte el valor lógico

# Ejemplos de operadores lógicos
x = True

y = False

print(x and y)  # False

print(x or y)   # True

print(not x)    # False
