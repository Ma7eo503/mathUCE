**Funciones: Utilidad, Tipos y Composición**

Las funciones son una herramienta fundamental en matemáticas y tienen una amplia variedad de aplicaciones en diferentes campos como la física, la ingeniería, la economía, y la informática.

# **Utilidad de las Funciones**

**Modelado Matemático:**

Las funciones permiten describir relaciones entre variables. Por ejemplo, la relación entre distancia y tiempo se puede modelar mediante una función en física.

**Análisis de Datos:**

En estadística y análisis de datos, las funciones se utilizan para modelar tendencias y realizar predicciones.

**Cálculo y Optimización:**

Las funciones son esenciales en el cálculo diferencial e integral, permitiendo encontrar máximos, mínimos y áreas bajo curvas.

**Programación y Algoritmos:**

En informática, las funciones permiten estructurar el código y facilitar la reutilización de bloques de código.

# **Según el Dominio y el Codominio:**

**Función Real:**

Mapea números reales en números reales.

**Función Compleja:**

Mapea números complejos en números complejos.

# **Según las Propiedades:**

**Función Inyectiva (Inyectiva o Uno a Uno):** Cada elemento del dominio mapea a un único elemento del codominio.

**Función Sobreyectiva (Sobre):** Cada elemento del codominio es imagen de al menos un elemento del dominio.

**Función Biyectiva:** Es inyectiva y sobreyectiva a la vez.
